```python
import pandas as pd
import numpy as np
import sklearn.preprocessing
```

```python
df = pd.read_csv('PersonDataSet.csv')
df
```

```python
gender_array = sklearn.preprocessing.LabelEncoder().fit_transform(np.array(df['gender']))
gender_array
```

```python
category_array = sklearn.preprocessing.LabelEncoder().fit_transform(np.array(df['category']))
category_array
```

```python
from_array = sklearn.preprocessing.OneHotEncoder().fit_transform(np.array(df['from']).reshape(-1,1)).toarray()
from_array
```

```python
print(gender_array.shape)
print(category_array.shape)
print(from_array.shape)
```

```python
data = np.concatenate((gender_array.reshape(-1,1),
                       category_array.reshape(-1,1),
                       from_array,
                       np.array(df['height']).reshape(-1,1),
                       np.array(df['weight']).reshape(-1,1),
                       np.array(df['age']).reshape(-1,1),
                       ),axis=1)
data
```

```python
pd.DataFrame(data).to_csv('个人数据集-处理后.csv', header=['gender',
                                     'category',
                                     'from_1','from_2','from_3',
                                     'height','weight','age'])
```
