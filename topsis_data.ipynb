```python
import pandas as pd
import re
import numpy as np
```

```python
df = pd.read_csv('data/销售类.csv', index_col=0)
# df = df.reset_index(drop=True)
df = df[['job', 'place', 'limit', 'tags', 'salary']]
df = df.reset_index(drop=True)
df
```

```python
for index, each in enumerate(df['salary']):
    print(each)
    loc = re.search(r'·\d+\w+', each)
    if loc is not None:
        df.loc[index, 'salary'] = str(each[:loc.span()[0]])
    if each.endswith('K'):
        df.loc[index, 'salary'] = str(float(each.split('-')[0]) + float(each.split('-')[1][:-1]) / 2)
    if each.endswith('元/天'):
        sum = 0
        if re.findall(r'\d+', each) is not None:
            for i in re.findall(r'\d+', each):
                sum += float(i)
            sum = sum * 4 * 30 / 1000
            df.loc[index, 'salary'] = str(sum / len(re.findall(r'\d+', each)))
```

```python
new_col = pd.DataFrame(np.random.normal(18, 4, len(df)),
                       columns=['working time'])
for index, each in enumerate(new_col['working time']):
    new_col.loc[index, 'working time'] = int(each)
```

```python
new_df = pd.concat([df, new_col], axis=1)
```
